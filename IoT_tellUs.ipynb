```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd 
import plotly.offline as py
import plotly.express as px
from mpl_toolkits.basemap import Basemap
%matplotlib inline



df_events = pd.read_csv("../input/tellus-sensors-data/Sensor_locations_in_Tellus.csv", dtype={'ID ': np.str})
df_events = df_events.rename(columns={"ID ": "ID"})
df_events = df_events[['ID','Locations','latitude','longitude','Serial number']]



# Load the train data and join on the events
df_train = pd.read_csv("../input/tellus-1000/saamplee.csv", dtype={'ID': np.str})
#df_train = df_train.rename(columns={"id": "Device ID"})
df_plot = pd.merge(df_train, df_events, on="ID", how="inner")



data = df_plot


fig = px.scatter_mapbox(data,
    lat='latitude',
    lon='longitude',
    color="light",
    size = "temperature",
    #color_continuous_scale="Viridis",
    animation_frame='timestamp',
    #hover_name ="Description",
    #hover_data = ["TOTAL_TAP_IN_VOLUME","BS_CODE"],
    #animation_frame="TIME_PER_HOUR",
    #category_orders={"TIME_PER_HOUR":Slider}
                       )

fig2 = (px.scatter_mapbox(data,
    lat='latitude',
    lon='longitude',
    color="light",
    size = "temperature",
    #color_continuous_scale="Viridis",
    #size_max=70, zoom=12,opacity=0.7,
    #hover_name ="Description",
   # hover_data = ["TOTAL_TAP_IN_VOLUME","BS_CODE"],
    #animation_frame='timestamp',
    #category_orders={"TIME_PER_HOUR":Slider}
                         ))
fig.add_trace(fig2.data[0])
fig.update(layout_coloraxis_showscale=True)
fig.update_layout(mapbox_style="carto-positron",
               mapbox_zoom=18,  mapbox_center = {"lat": 65.058872, "lon": 25.466510})
fig.update_layout(margin={"r":0,"t":30,"l":0,"b":0})
fig.show() 
```
